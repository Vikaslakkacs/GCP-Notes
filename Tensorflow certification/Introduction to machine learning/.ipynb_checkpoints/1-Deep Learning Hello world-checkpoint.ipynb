### lets work on some basic example

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

## Lets create input data and answers and get the rules

```python
x=np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=np.float16)
y=np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=np.float16)
```

```python
x
```

#### Now letscreate a model and train the data

```python
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
```

```python
model.summary()
```

### Lets compile the model
#### For compiling we have to provide optimizer and loss function

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

### Now we will train the data

```python
model.fit(x,y,epochs=500)
```

### After trainig we will predict themodel with some value which the model has not seen

```python
model.predict([10.0])
```

### why did not it give the exact result.
#### Because it runs on probability.
