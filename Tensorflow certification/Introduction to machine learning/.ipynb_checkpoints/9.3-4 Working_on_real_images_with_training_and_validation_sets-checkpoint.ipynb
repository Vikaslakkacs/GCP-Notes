```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
image_data_gen=ImageDataGenerator(rescale=1/255)
```

```
training_path='/tmp/horse-or-human'
training_data_gen=image_data_gen.flow_from_directory(training_path,batch_size=128,target_size=(300,300),
                                                     class_mode='binary',shuffle=True
                                )
```

```
test_path='/tmp/validation-horse-or-human'
test_data_gen=image_data_gen.flow_from_directory(test_path,target_size=(300,300),class_mode='binary',batch_size=64,shuffle=False)
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
```

```
model=keras.Sequential()
model.add(Conv2D(filters=16,kernel_size=(3,3),strides=(1,1),padding='SAME',input_shape=(300,300,3),activation=tf.nn.relu))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),padding='SAME',activation=tf.nn.relu))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),padding='SAME',activation=tf.nn.relu))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(units=512,activation=tf.nn.relu))
model.add(Dense(units=1,activation=tf.nn.sigmoid))
model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=0.0001),loss=keras.losses.binary_crossentropy,metrics=['accuracy'])
model.summary()
```

```
model.fit_generator(training_data_gen,
                    steps_per_epoch=8,
                    epochs=15,verbose=2,
                    validation_data=test_data_gen,validation_steps=8)
```
