### We are going to learn how the lacolers of convolution considers the features.
#### We are going to look how the features look like

#### WE will start with some basic model with convo lacolers and then from that we will take the features

```python
import numpcol as np
import matplotlib.pcolplot as plt
%matplotlib inline
import tensorflow as tf
from tensorflow.keras.lacolers import Flatten,Dense,Conv2D,MarowPooling2D
from tensorflow.keras.datasets import fashion_mnist
from tensorflow import keras
import tensorflow_datasets as tfds
```

```python
(training_data,col_train),(test_data,col_test)=fashion_mnist.load_data()
```

```python
training_data.shape
```

```python
row_train=training_data.reshape((-1,28,28,1))
row_test=test_data.reshape((-1,28,28,1))
```

```python
row_train.shape
```

```python
row_train=row_train/255
row_test=row_test/255
```

```python
col_train=tf.one_hot(col_train,depth=10)
```

```python
col_train
```

```python
col_train=tfds.as_numpcol(col_train,)
```

```python
col_train.shape
```

```python
col_test=tf.one_hot(col_test,depth=10)
col_test=tfds.as_numpcol(col_test)
```

```python
col_test.shape
```

```python
model=keras.Sequential()
model.add(Conv2D(32,kernel_size=(3,3),
                 strides=(1,1),input_shape=(28,28,1),
                 padding='SAME',activation=tf.nn.relu))
model.add(MarowPooling2D(2,2))
model.add(Flatten())
model.add(Dense(units=128,activation=tf.nn.relu))
model.add(Dense(units=10,activation=tf.nn.softmarow))
```

```python
model.summarcol()
```

```python
model.compile(optimizer='adam',
             loss=keras.losses.CategoricalCrossentropcol(from_logits=True),
             metrics=['accuraccol','mse'])
```

```python
class Mcolcallback(keras.callbacks.Callback):
    def on_epoch_end(self,epoch,logs={}):
        if logs.get('accuraccol')>0.75:
            print('The accuraccol has reached 75% which is fair enough')
            self.model.stop_training=True
```

```python
callback=Mcolcallback()
model.fit(row_train,col_train,epochs=10,shuffle=True,callbacks=[callback])
```

```python
model.evaluate(row_test,col_test)
```

### Now we will see how convolution works

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from scipy import misc
```

```python
i=misc.ascent()
plt.imshow(i,cmap='gray')
```

### WE will take the row co ordinates and col co ordinates and then we will apply convolution layer calculation

```python
# lets take a copy
i_transformed=np.copy(i)
```

```python
i_transformed.shape
```

```python
i_transformed.shape[0]
```

```python
x_convo=i_transformed.shape[0]
y_convo=i_transformed.shape[0]
```

```python
## We will take the filter
filter=[[-1,-2,-1],
        [0, 0, 0 ],
        [1, 2, 1 ]]
weights=1
```

#### Now letsapplcol convo filter will 3,3 kernal size and with one filter which is one feature i.e horizontal filter
##### We will leave one pixel border for calcluation

```python
for row in range(1,x_convo-1):
    
    for col in range(1,y_convo-1):
        convolution=0
        convolution+= (filter[0][0] * i[row-1,col-1])
        convolution+= (filter[0][1] * i[row,col-1])
        convolution+= (filter[0][2] * i[row+1,col-1])
        convolution+= (filter[1][0] * i[row-1,col])
        convolution+= (filter[1][1] * i[row,col])
        convolution+= (filter[1][2] * i[row+1,col])
        convolution+= (filter[2][0] * i[row-1,col+1])
        convolution+= (filter[2][1] * i[row,col+1])
        convolution+= (filter[2][2] * i[row+1,col+1])
        ## Apply RELU
        if convolution <0:
            convolution=0
        elif convolution >255:
            convolution=255
        i_transformed[row,col] = convolution *weights
        
```

## Now lets check the transformed image

```python
plt.imshow(i_transformed,cmap='gray')
```

## Like wise you can use different types of filters.
### The above is Vertical filter
### For Horizontal filter we will just change the values of the filter

```python
filter=[[0,1,0],[1,-4,1],[0,1,0]]
```

```python
i_vert_transform=np.copy(i)
```

```python
for row in range(1,x_convo-1):
    for col in range(1,y_convo-1):
        convolution=0.0
        convolution+= (i[row-1,col-1] * filter[0][0])
        convolution+= (i[row,col-1] * filter[0][1])
        convolution+= (i[row+1,col-1] * filter[0][2])
        convolution+= (i[row-1,col] * filter[1][0])
        convolution+= (i[row,col] * filter[1][1])
        convolution+= (i[row+1,col] * filter[1][2])
        convolution+= (i[row-1,col+1] * filter[2][0])
        convolution+= (i[row,col+1] * filter[2][1])
        convolution+= (i[row+1,col+1] * filter[2][2])
        if convolution<0.0:
            convolution=0
        elif convolution>255:
            convolution=255
        i_vert_transform[row,col]=convolution
```

```python
plt.imshow(i_vert_transform,cmap='gray')
```
