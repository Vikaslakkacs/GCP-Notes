#### Now we have got the images. Lets start training the model.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense,Conv2D,Flatten,MaxPooling2D
```

```python
model=keras.Sequential()
model.add(Conv2D(16,kernel_size=(3,3),
                 strides=(1,1),input_shape=(300,300,3),
                 activation='relu',padding='SAME'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(32,kernel_size=(3,3),
                strides=(1,1),activation='relu',
                padding='SAME'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64,kernel_size=(3,3),
                strides=(1,1),activation='relu',
                padding='SAME'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(units=512,activation='relu'))
## for binary classification we use sigmoid
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=0.0001),
             loss=keras.losses.binary_crossentropy,
             metrics=['accuracy'])
```

### We have created model NOw we have to prepare the data.
#### We dnt have images in some dataset library. we have it in our folder.
#### We will create the path using Imagedatagenerator module. Which will create data with desired shape, batch size and much more parameters.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
image_data_gen=ImageDataGenerator(rescale=1/255)
```

```python
directory_path=os.path.join('/tmp/horse-or-human')
```

```python
train_gen=image_data_gen.flow_from_directory(directory=directory_path,
                                        target_size=(300,300),batch_size=128,
                                            class_mode='binary')
```

```python
train_gen
```

### Now lets train the model

```python
model.fit(train_gen,
         steps_per_epoch=8,
          epochs=15,
          verbose=2
         )
```
