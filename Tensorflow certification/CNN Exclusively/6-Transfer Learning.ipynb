```python
### we are going to learn transfer learning
### Transfer learning: We can take already created model and to that model we customize it by adding layers and training
#### This is useful because we may not have huge data sets and high computing power to traing and develop the model. So in this case transfer learning is more useful
#### Another advantage is, we can use already optimized model and can utilize it for our project
```

```python
### We are downloading the h5. file of inception which is a power full image classification which is trained on more than 10 million images with more classifiers
!wget --no-check-certificate \
    https://storage.googleapis.com/mledu-datasets/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5 \
    -O /tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5
```

```python
#### now we will create model and then add weights to the existing model.
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.models import Model
from tensorflow.keras import layers
```

```python
##Now we got the h5 file. So lets load the weights
### include_top is a boolean whether to connect with fully connected layer or not. Here we are not including it and we wil create FC on our own so False.
### WEights: Asking whether to include weights or not.
pre_trained_model=InceptionV3(include_top=False,input_shape=(200,200,3),weights=None)
```

```python
### We will load weights from h5 file
pre_trained_model.load_weights('/tmp/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5')
```

```python
### Now we have inception model with all the layers except FC layers as we are going to develop.
pre_trained_model.layers
```

```python
### While training we dnt need all the layers to learn thedata, Only FC layers are enough to learn.
## So we have option that we can disable training, Which is we can turn off traininig for layers.
for layer in pre_trained_model.layers:
      layer.trainable=False
```

```python
### now to connect out customized FC layer we have to take the last layer and connect it.
## last layer is mixed7, so we will get that layer and consider the output and then customize FC
last_layer=pre_trained_model.get_layer('mixed7')
```

```python
last_layer_output=last_layer.output
```

```python
last_layer_output
```

```python
### Now we have got enough data lets start building customizer model
x=layers.Flatten()(last_layer_output)
x=layers.Dense(units=1024,activation=tf.nn.relu)(x)
x=layers.Dense(units=1,activation=tf.nn.sigmoid)(x)

##3 we have got the layers now lets build model
model=Model(inputs=pre_trained_model.input,outputs=x)
```

```python
model.compile(optimizer='RMSprop',loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
### We have got the model 
## Now we will load the data and train
!wget --no-check-certificate \
        https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip \
       -O /tmp/cats_and_dogs_filtered.zip
```

```python
import os
import zipfile
zip_file=zipfile.ZipFile('/tmp/cats_and_dogs_filtered.zip','r')
zip_file.extractall('/tmp/cats_dogs')
zip_file.close()
```

```python
path=os.path.join('/tmp/cats_dogs','cats_and_dogs_filtered')
```

```python
train_path=os.path.join(path,'train')
test_path=os.path.join(path,'validation')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_data_gen=ImageDataGenerator(rescale=1/255.0,
                                  rotation_range=40,width_shift_range=0.2,height_shift_range=0.2,
                                  shear_range=0.2,zoom_range=0.2,
                                  fill_mode='nearest',horizontal_flip=True)
train_gen=train_data_gen.flow_from_directory(train_path,target_size=(200,200),
                                             class_mode='binary',
                                             batch_size=20,shuffle=True)
test_data_gen=ImageDataGenerator(rescale=1/255.0)
test_gen=test_data_gen.flow_from_directory(test_path,target_size=(200,200),
                                           class_mode='binary',batch_size=20)
```

```python
model.fit(train_gen,steps_per_epoch=100,epochs=10,validation_data=test_gen,validation_steps=50)
```

```python
metrics=model.history.history
loss=metrics['loss']
val_loss=metrics['val_loss']
```

```python
acc=metrics['accuracy']
val_acc=metrics['val_accuracy']
```

```python
plt.plot(loss)
plt.plot(val_loss)
```

```python
plt.plot(acc)
plt.plot(val_acc)
```
